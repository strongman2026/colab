```python
# cell4: Flux 2D Assets 2026 V12（手工按需运行，aria2多路 + 实时进度）
import os, json, zipfile, subprocess
from pathlib import Path
from google.colab import userdata

DATASET_REF = "jarredstroman/flux-2d-assets-2026-v12"
WORK_DIR = Path("/content/dataset")
ZIP_PATH = WORK_DIR / "flux-2d-assets-2026-v12.zip"
COMFY_DIR = Path("/content/ComfyUI")

def run(cmd, desc="", check=True):
    if desc:
        print(f"\n[⏳] {desc}")
    print(f"[CMD] {cmd}")
    p = subprocess.run(cmd, shell=True, text=True, stdout=subprocess.PIPE, stderr=subprocess.STDOUT)
    out = p.stdout or ""
    print(out[-1800:] if len(out) > 1800 else out)
    if check and p.returncode != 0:
        raise RuntimeError(f"命令失败: {cmd}")
    return p

def setup_kaggle_auth():
    u = userdata.get("KAGGLE_USERNAME")
    k = userdata.get("KAGGLE_KEY")
    kg = Path("/root/.kaggle")
    kg.mkdir(parents=True, exist_ok=True)
    j = kg / "kaggle.json"
    j.write_text(json.dumps({"username": u, "key": k}), encoding="utf-8")
    os.chmod(j, 0o600)
    print("✅ Kaggle 凭据写入完成")

def get_download_url_via_python_api(dataset_ref):
    # 用官方 Python API 拿真实下载链接
    from kaggle.api.kaggle_api_extended import KaggleApi
    api = KaggleApi()
    api.authenticate()

    # dataset_ref = owner/slug
    owner, slug = dataset_ref.split("/", 1)
    # 返回 Dataset 对象，含 ref / slug / owner 等
    ds_list = api.dataset_list(search=slug, user=owner)
    if not ds_list:
        raise RuntimeError(f"❌ 找不到数据集: {dataset_ref}")
    # 精确匹配
    target = None
    for d in ds_list:
        # d.ref 形如 owner/slug
        if getattr(d, "ref", "") == dataset_ref:
            target = d
            break
    if target is None:
        # 兜底取第一条
        target = ds_list[0]

    # 直接调用底层client拿下载URL（与CLI本质一致）
    # endpoint: datasets/download/{owner}/{slug}
    url = f"https://www.kaggle.com/api/v1/datasets/download/{owner}/{slug}"

    # 用授权头请求并跟随重定向拿最终可下载链接
    import requests

    # 构造 basic auth
    username = userdata.get("KAGGLE_USERNAME")
    key = userdata.get("KAGGLE_KEY")
    r = requests.get(url, auth=(username, key), allow_redirects=False)

    if r.status_code in (301, 302, 303, 307, 308):
        loc = r.headers.get("Location")
        if not loc:
            raise RuntimeError("❌ 未获取到重定向下载链接")
        return loc
    elif r.status_code == 200:
        # 少见，直接可下载
        return url
    else:
        raise RuntimeError(f"❌ 获取下载链接失败: HTTP {r.status_code}, body={r.text[:300]}")

def aria2_download(url, zip_path: Path):
    zip_path.parent.mkdir(parents=True, exist_ok=True)
    cmd = (
        f'aria2c -c -x 16 -s 16 -k 1M '
        f'--summary-interval=1 --download-result=full '
        f'--auto-file-renaming=false '
        f'--out "{zip_path.name}" -d "{zip_path.parent}" '
        f'"{url}"'
    )
    run(cmd, "aria2 多路下载（显示实时速度/进度）")

def unzip_to_workdir(zip_path: Path, work_dir: Path):
    print("\n[⏳] 解压数据集")
    with zipfile.ZipFile(zip_path, "r") as zf:
        zf.extractall(work_dir)
    print("✅ 解压完成")

def find_flux_model(root: Path):
    prefer = [
        root / "flux1-schnell.safetensors",
        root / "flux1-dev.safetensors",
        root / "flux1-schnell-fp8.safetensors",
    ]
    for p in prefer:
        if p.exists():
            return p
    all_sf = list(root.rglob("*.safetensors"))
    if not all_sf:
        return None
    for p in all_sf:
        if "schnell" in p.name.lower():
            return p
    return all_sf[0]

def link_to_comfy(src: Path):
    unet = COMFY_DIR / "models/unet/flux1-schnell.safetensors"
    ckpt = COMFY_DIR / "models/checkpoints/flux1-schnell.safetensors"
    unet.parent.mkdir(parents=True, exist_ok=True)
    ckpt.parent.mkdir(parents=True, exist_ok=True)
    for dst in [unet, ckpt]:
        if dst.exists() or dst.is_symlink():
            try: dst.unlink()
            except: pass
        os.symlink(str(src), str(dst))
        print(f"✅ {dst} -> {src}")

print("\n================ [1/6] 安装依赖 ================")
run("python -m pip install -q -U kaggle", "安装 kaggle")
run("apt-get update -qq && apt-get install -y -qq aria2", "安装 aria2")

print("\n================ [2/6] Kaggle 鉴权 ================")
setup_kaggle_auth()

print("\n================ [3/6] 获取下载链接 ================")
dl_url = get_download_url_via_python_api(DATASET_REF)
print("✅ 获取下载链接成功（已隐藏URL）")

print("\n================ [4/6] aria2 多路下载 ================")
aria2_download(dl_url, ZIP_PATH)
if not ZIP_PATH.exists():
    raise RuntimeError("❌ 下载完成后未找到 zip 文件")

print("\n================ [5/6] 解压并定位模型 ================")
unzip_to_workdir(ZIP_PATH, WORK_DIR)
src = find_flux_model(WORK_DIR)
if not src:
    raise RuntimeError("❌ 未在数据集中找到 .safetensors")
print(f"✅ 选定模型: {src}")

print("\n================ [6/6] 挂载到 ComfyUI ================")
link_to_comfy(src)

print("\n🎉 完成：Flux 模型已就绪，运行启动 cell 即可。")
```
